```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


def VisualizeHeatmap(cm):
    ax = plt.subplot()
    sns.set(font_scale=1.5)
    sns.heatmap(cm, cmap="Greens", annot=True, cbar=False, square=True, ax=ax);
    
    # Labels, title and ticks
    label_font = {"size":"15"}
    ax.set_xlabel("Predicted label", fontdict=label_font);
    ax.set_ylabel("True label", fontdict=label_font);
    
    ax.tick_params(axis='x', which="major", labelsize=15)
    ax.tick_params(axis='y', which="major", labelsize=15, labelrotation=0)
    ax.xaxis.set_ticklabels(["AN", "DI", "FE", "HA", "NE", "SA", "SU"]);
    ax.yaxis.set_ticklabels(["AN", "DI", "FE", "HA", "NE", "SA", "SU"]);

    for spine in ax.spines.values():
        spine.set(visible=True, lw=.8, edgecolor="black")
    
    plt.show()

cm_vanilla = np.array(
    [[4, 0, 0, 0, 0, 2, 0],
     [0, 2, 0, 0, 1, 3, 0],
     [0, 0, 3, 0, 1, 2, 1],
     [0, 0, 0, 6, 0, 0, 0],
     [0, 0, 0, 2, 3, 0, 1],
     [0, 0, 0, 1, 0, 5, 0],
     [0, 0, 0, 0, 1, 0, 5]])

cm_proposed = np.array(
    [[6, 0, 0, 0, 0, 0, 0],
     [1, 5, 0, 0, 0, 0, 0],
     [0, 0, 5, 0, 0, 1, 1],
     [0, 0, 0, 6, 0, 0, 0],
     [0, 0, 0, 0, 6, 0, 0],
     [0, 0, 1, 1, 0, 4, 0],
     [0, 0, 0, 0, 1, 0, 5]])


# VisualizeHeatmap(cm_vanilla)
VisualizeHeatmap(cm_proposed)
```
